```python
from callAPI import CallAPI
from gcloud import UploadGCloud
```

```python
api = CallAPI()
url = "http://asterank.com/api/mpc?query={}&limit=10000"

request = api.getJSON(url)
api.saveLocal(request, "data1.json", "data")

cloud = UploadGCloud()

blob = "160425_.json"
bucket_name = "example_bucket_1304"
cloud.upload_nolocal(bucket_name, request, blob)
```
